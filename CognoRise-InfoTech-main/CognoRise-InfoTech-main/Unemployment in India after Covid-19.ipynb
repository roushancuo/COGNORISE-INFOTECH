#         **Unemployment in India**

**Context**

The story behind this datasets is how lock-down affects employment opportunities and how the unemployment rate increases during the Covid-19.


This dataset contains the unemployment rate of all the states in India.

Region = states in India

Date = date which the unemployment rate observed

Frequency = measuring frequency (Monthly)

Estimated Unemployment Rate (%) = percentage of people unemployed in each States of India

Estimated Employed = percentage of people employed

Estimated Labour Participation Rate (%) = labour force participation rate by dividing the number of people actively participating in the labour force by the
total number of people eligible to participate in the labor force


1. Problem Description

The metric used to gauge unemployment is the unemployment rate, representing the proportion of unemployed individuals relative to the total labor force. The COVID-19 pandemic has led to a significant surge in the unemployment rate. This analysis aims to explore the socio-economic repercussions of the pandemic on India's workforce and labor market.

The dataset facilitates an understanding of the variations in unemployment trends among different states in India during the COVID-19 crisis. It provides valuable information on the effects on the unemployment rate, employment statistics, and labor participation rates in diverse regions of the country.

```python
df = df.drop([' Frequency'],axis = 1)
```

```python
# Renaming columns for easier access
df1= df.rename(columns={ ' Estimated Unemployment Rate (%)' : 'est_unemp_perc', ' Estimated Employed' : 'est_emp',
                          ' Estimated Labour Participation Rate (%)' : 'est_labour_perc'}).reset_index(drop = True)
```

```python
df1
```

```python
df1.isna().sum()
```

```python
df1[' Date'] = pd.to_datetime(df1[' Date'])
```

```python
import datetime as dt  #commonly used for working with dates and times.
```

```python
df1[' Date']=df1[' Date'].fillna((df1[' Date']).mean())
df1['Region']=df1['Region'].fillna((df1['Region']).mode()[0])
df1['est_unemp_perc']=df1['est_unemp_perc'].fillna((df1['est_unemp_perc']).mean())
df1['est_emp']=df1['est_emp'].fillna((df1['est_emp']).mean())
df1['est_labour_perc']=df1['est_labour_perc'].fillna((df1['est_labour_perc']).mean())
df1['Area']=df1['Area'].fillna((df1['Area']).mode()[0])
```

```python
df1.corr()
```

```python
sns.heatmap(df1.corr(),annot= True)
```

```python
plt.title('Unemployment Rate based on Area')
sns.scatterplot(y=df1['Region'],x=df1['est_unemp_perc'],hue=df1['Area'])
```

**Data Visualization**

```python
df1[' Date'] = pd.to_datetime(df1[' Date'])
#converted the 'Date' column to a datetime
```

```python
df2 = df1.groupby('Region')[['est_unemp_perc']].mean().sort_values(by='est_unemp_perc',ascending= False)
df2.plot(kind='bar')             #used to create a bar plot of the data in a DataFrame
plt.title('Estimated unemployment percentage per region')
plt.ylabel('Unemployment percentage')
plt.xticks(rotation= 90)
plt.figtext(x= 0.19, y= 0.7, s= 'Tripura has the highest\nunemloyment percentage\nof all regions')
plt.show()
```


The average employed per region will tell us if there a significant difference in employees among states as well as the regions with the highest and lowest number of employees.


```python
df2 = df1.groupby('Region')[['est_emp']].mean().sort_values(by='est_emp',ascending= False)
df2.plot(kind='bar')
plt.title('Average millions employed per region')
plt.ylabel('Millions employed')
plt.figtext(x=0.15, y=0.75, s='While the numbers of employees have\na big variance among regions,\nUttar Pradesh has the most employees')
plt.show()
```

```python
df2= df1.groupby('Region')['est_unemp_perc'].agg(lambda x: max(x) - min(x)).sort_values(ascending=False).plot(kind='bar')
plt.suptitle('The difference in unemployment rate per region.')
plt.title('Maximum rate - Minimum rate')
plt.figtext(x= 0.15, y= 0.75, s='Puducherry is the most affected\nby the crisis with more then\n70 percent difference in unemployment rate ')
plt.show()
```
